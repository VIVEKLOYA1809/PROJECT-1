```python
import pandas as pd
import numpy as np
```

```python
skill=pd.read_excel("skill_data.xlsx")
x=skill["Skills"].values
```

```python
web_developer_skills=['django','html','css','nodejs','reactjs','nodejs','flask','laravel']
machine_learning_skills=['nosql','pandas','numpy','matpotlib','seabron','scipy']
```

```python
sk=[]
length=[]
for i in range(len(skill)):
    l=[str[1:-1] for str in x[i][1:-1].split(", ")]
    p=np.intersect1d(l,web_developer_skills)
    q=np.intersect1d(l,machine_learning_skills)
    length.append(len(np.unique(l)))
    if len(p)>=3:
        sk.append("web_development")
    elif len(q)>=3:
        sk.append("machine_learning")
    else:
        sk.append("No_Skill")
        
    
```

```python
skill["Designation"]=sk
skill["No_of_Skills"]=length
```
