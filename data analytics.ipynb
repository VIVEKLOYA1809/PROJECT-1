```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_excel("Train_data.xlsx",header=None)
```

```python
dataset_North=dataset.iloc[1:234,3:]
dataset_North=dataset_North.T
dataset_North.columns=dataset.iloc[1:234,2].values
dataset_North.index=dataset.iloc[0,3:].values
```

```python
outliers={}
def detect_outliers(data):
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        z_score=(i-mean)/std
        if (z_score>1.9) or (z_score<-1.5) :
            outliers.update({data.index(i):i})
    return outliers
```

```python
def moving_average(data,i):

    # for i in len(data):

    if(i == 0):
        avg = (data[i]+data[i+1]+data[i+2] ) / 3.0

    elif(i == len(data)-1):
        avg = (data[i-2]+data[i-1]+data[i] ) / 3.0

    else:
        avg = (data[i-1] + data[i] + data[i+1]) / 3.0
    
    return avg
```

```python
dataset_North_without_outliers=dataset_North.copy()
dataset_North_without_outliers=dataset_North_without_outliers.astype(float)
```

```python
out_North={}
for i in range(dataset_North.shape[1]):
    outliers={}
    out_North.update({dataset_North.columns[i]:detect_outliers(list(dataset_North.iloc[:,i].values))})
out_North
for items,value in out_North.items():
    for i,j in value.items():
        dataset_North_without_outliers[items][i]=round(moving_average(dataset_North[items],i))
```

```python
def MAPE(Y_actual,Y_pred):
    mape = np.abs((Y_actual - Y_pred)/Y_actual)
    return mape
```

```python
w=[]
from statsmodels.tsa.arima.model import ARIMA
from pmdarima import auto_arima
for i in range(20):
  stepwise_fit=auto_arima(dataset_North_without_outliers.iloc[:37,i].values,trace=True)
  a=int(input("first"))
  b=int(input("second"))
  c=int(input("third"))
  model=ARIMA(dataset_North_without_outliers.iloc[:37,0].values,order=(a,b,c))
  model=model.fit()
  Y_pred=model.predict(start=37,end=37,type='levels')
  Y_actual=dataset_North_without_outliers.iloc[37,0]
  w.append(MAPE(Y_pred,Y_actual))
```
